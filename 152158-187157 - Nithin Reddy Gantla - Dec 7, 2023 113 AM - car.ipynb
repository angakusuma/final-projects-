```python
### 1. Deriving New Features From "Company Name" Feature.
```

```python
z = round(df.groupby(["CompanyName"])["price"].agg(["mean"]),2).T
z
```

```python
df = df.merge(z.T,how="left",on="CompanyName")
```

```python
bins = [0,10000,20000,40000]
cars_bin=['Budget','Medium','Highend']
df['CarRange'] = pd.cut(df['mean'],bins,right=False,labels=cars_bin)
df.head()
```

```python
### 1. Creating new DataFrame with all the useful Features.
```

```python
new_df = df[['fueltype','aspiration','doornumber','carbody','drivewheel','enginetype','cylindernumber','fuelsystem'
             ,'wheelbase','carlength','carwidth','curbweight','enginesize','boreratio','horsepower','citympg','highwaympg',
             'price','CarRange']]
```

```python
new_df.head()
```

```python
### 2. Creating Dummies Variables for all the Categorical Features.
```

```python
new_df = pd.get_dummies(columns=["fueltype","aspiration","doornumber","carbody","drivewheel","enginetype",
                                "cylindernumber","fuelsystem","CarRange"],data=new_df)
```

```python
new_df.head()
```

```python
### 3. Feature Scaling of Numerical Data.
```

```python
scaler = StandardScaler()
```

```python
num_cols = ['wheelbase','carlength','carwidth','curbweight','enginesize','boreratio','horsepower',
            'citympg','highwaympg']

new_df[num_cols] = scaler.fit_transform(new_df[num_cols])
```

```python
new_df.head()
```

```python
### 4. Selecting Features & Labels for Model Training & Testing.
```

```python
x = new_df.drop(columns=["price"])
y = new_df["price"]
```

```python
x.shape
```

```python
y.shape
```

```python
### 5. Splitting Data for Model Traning & Testing.
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("x_train - >  ",x_train.shape)
print("x_test - >  ",x_test.shape)
print("y_train - >  ",y_train.shape)
print("y_test - >  ",y_test.shape)
```

```python
training_score = []
testing_score = []
```

```python
training_score_r2 = []
testing_score_r2 = []
training_score_mse = []
testing_score_mse = []
training_score_mae = []
testing_score_mae = []

def model_prediction(model):
    model.fit(x_train, y_train)
    x_train_pred = model.predict(x_train)
    x_test_pred = model.predict(x_test)
    
    # R-squared score
    r2_train = r2_score(y_train, x_train_pred)
    r2_test = r2_score(y_test, x_test_pred)
    
    # Mean Squared Error (MSE)
    mse_train = mean_squared_error(y_train, x_train_pred)
    mse_test = mean_squared_error(y_test, x_test_pred)
    
    # Mean Absolute Error (MAE)
    mae_train = mean_absolute_error(y_train, x_train_pred)
    mae_test = mean_absolute_error(y_test, x_test_pred)

    training_score_r2.append(r2_train)
    testing_score_r2.append(r2_test)
    training_score_mse.append(mse_train)
    testing_score_mse.append(mse_test)
    training_score_mae.append(mae_train)
    testing_score_mae.append(mae_test)

    print(f"R-squared  {model} model Training Data is:", r2_train)
    print(f"R-squared  {model} model Testing Data is:", r2_test)
    
    print(f"Mean Absolute Error {model} model on Training Data is:", mae_train)
    print(f"Mean Absolute Error {model} model on Testing Data is:", mae_test)


linear_reg_model = LinearRegression()
model_prediction(linear_reg_model)
```

```python
model_prediction(DecisionTreeRegressor())
```
